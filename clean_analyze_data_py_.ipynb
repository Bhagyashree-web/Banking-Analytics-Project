```python
import pandas as pd
from datetime import datetime

# Load sample data if not already loaded
customers = pd.read_csv("customer.csv")     # Must include columns: customer_id, dob, income, ...
loans = pd.read_csv("loan.csv")             # Must include columns: loan_type, loan_amount, customer_id

# Clean column names (optional, but recommended)
customers.columns = customers.columns.str.strip().str.lower().str.replace(' ', '_')
loans.columns = loans.columns.str.strip().str.lower().str.replace(' ', '_')

# ➤ 1. Calculate age from date of birth
# Add a dummy dob column
customers['dob'] = ['1990-05-12', '1985-11-23', '2000-08-30']  # Adjust to match number of rows

# Now calculate age
customers['age'] = pd.to_datetime('today').year - pd.to_datetime(customers['dob']).dt.year

# ➤ 2. Total loan amount by type
loan_summary = loans.groupby('loan_type')['loan_amount'].sum().reset_index()
print("📊 Loan Summary by Type:")
display(loan_summary)

# ➤ 3. Merge loan and customer data
merged = pd.merge(loans, customers, on='customer_id')
print("🔗 Merged Data (Loans + Customers):")
display(merged.head())

# ➤ 4. Filter high income customers
customers['income'] = [900000, 700000, 850000]  # match number of rows

# Now filter high-income customers
high_income = customers[customers['income'] > 800000]
print(high_income)


```

```python
# Step 1: Import Required Libraries
import pandas as pd

# Step 2: Load CSV Files (Make sure all these files are uploaded in Colab)
df_customer = pd.read_csv("customer.csv")
df_account = pd.read_csv("account.csv")
df_loan = pd.read_csv("loan.csv")
df_transaction = pd.read_csv("transaction.csv")
df_marketing = pd.read_csv("marketing.csv")

# Step 3: Clean Column Names (Optional but recommended)
for df in [df_customer, df_account, df_loan, df_transaction, df_marketing]:
    df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Step 4: Calculate Age if 'dob' column is present in customer data
if 'dob' in df_customer.columns:
    df_customer['age'] = pd.to_datetime('today').year - pd.to_datetime(df_customer['dob']).dt.year

# Step 5: Print Sample Data
print("📋 Customers:")
display(df_customer.head())

print("\n🏦 Accounts:")
display(df_account.head())

print("\n💰 Loans:")
display(df_loan.head())

print("\n🔄 Transactions:")
display(df_transaction.head())

print("\n📣 Marketing Campaigns:")
display(df_marketing.head())

# Step 6: Summary Examples

# a. Total Loan Amount by Type
loan_summary = df_loan.groupby('loan_type')['loan_amount'].sum().reset_index()
print("\n📊 Total Loan Amount by Type:")
display(loan_summary)

# b. Customer Participation in Campaigns
campaign_response = df_marketing[df_marketing['response'].str.lower() == 'yes']
print("\n📬 Customers Who Responded YES:")
display(campaign_response.merge(df_customer, on='customer_id')[['name', 'campaign_name', 'response']])
```
