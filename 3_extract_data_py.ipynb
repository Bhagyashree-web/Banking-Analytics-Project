```python
import pandas as pd

# Load your CSVs into DataFrames
df_customer = pd.read_csv("customer.csv")
df_account = pd.read_csv("account.csv")
df_loan = pd.read_csv("loan.csv")
df_transaction = pd.read_csv("transaction.csv")
df_marketing = pd.read_csv("marketing.csv")
```

```python
# Check first few rows
print(df_customer.head())
print(df_account.head())
print(df_loan.head())
print(df_transaction.head())
print(df_marketing.head())
```

```python
yes_responses = df_marketing[df_marketing['response'] == 'Yes']
result1 = pd.merge(yes_responses, df_customer, on='customer_id')[['name', 'campaign_name', 'response']]
print(result1)
```

```python
merged = pd.merge(yes_responses, df_customer, on='customer_id')
loan_merged = pd.merge(merged, df_loan[df_loan['status'] == 'Active'], on='customer_id')
print(loan_merged[['name', 'campaign_name', 'loan_type', 'status']])
```
